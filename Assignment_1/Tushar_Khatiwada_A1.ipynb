#### Question 6: List Sanitization and Metrics
Given:
```python
marks = [45, -3, 88, "NA", 102, None, 67]
```

Write a function that:
1. Removes invalid entries  
2. Returns:
   - Cleaned list  
   - Number of invalid entries  
   - Percentage of valid marks  

Return values using a tuple.



````python
marks = [45, -3, 88, "NA", 102, None, 67]

valid_marks = []
invalid_count = 0

for m in marks:
    if type(m) == int or type(m) == float:
        if m >= 0 and m <= 100:
            valid_marks.append(m)
        else:
            invalid_count += 1
    else:
        invalid_count += 1

valid_percentage = (len(valid_marks) / len(marks)) * 100

print("Cleaned List:", valid_marks)
print("Invalid Entries:", invalid_count)
print("Valid Marks Percentage:", valid_percentage)
````

#### Question 9: Function Composition
Write functions to:
1. Calculate total marks  
2. Calculate average  
3. Classify result (Distinction / Pass / Fail)

Then write a function that combines them and returns:
```
(name, total, average, result)
```

Explain why modular design is beneficial.


````python
# Function to calculate the total marks
# It takes a list of marks and returns their sum
def calculate_total(marks):
    return sum(marks)


# Function to calculate the average marks
# It takes total marks and number of subjects
def calculate_average(total, number_of_subjects):
    return total / number_of_subjects


# Function to classify the result based on average
# Distinction: average >= 75
# Pass: average >= 40
# Fail: average < 40
def classify_result(average):
    if average >= 75:
        return "Distinction"
    elif average >= 40:
        return "Pass"
    else:
        return "Fail"


# Main function that combines all calculations
# It returns name, total marks, average, and result
def student_result(name, marks):
    total = calculate_total(marks)              # Calculate total marks
    average = calculate_average(total, len(marks))  # Calculate average
    result = classify_result(average)           # Decide result

    return (name, total, average, result)

marks = [88, 85, 90, 96, 96]   # Marks obtained in different subjects
output = student_result("Tushar", marks)
print(output)
````

#### Question 10: Defensive Input Parsing
Write a function that accepts a raw input string in the format:

```
"Ram|101|22|78, 88, 91"
```
Hint: Parse student data from string format: "Name|Roll|Age|Mark1, Mark2, Mark3"

Tasks:
1. Extract name, roll number, age, and marks  
2. Convert them into appropriate data types  
3. Validate age (≥16) and marks (0–100)  
4. Return the result as a dictionary using roll number as the key  

The function must not crash for malformed input.


````python
def parse_student_record(data):
    result = {}

    try:
        # Split the input string
        parts = data.split("|")
        name = parts[0]
        roll = int(parts[1])
        age = int(parts[2])

        # Age validation
        if age < 16:
            return {}

        # Split and convert marks
        marks = []
        for m in parts[3].split(","):
            mark = int(m.strip())
            if mark < 0 or mark > 100:
                return {}
            marks.append(mark)

        # Store result using roll number as key
        result[roll] = {
            "Name": name,
            "Age": age,
            "Marks": marks
        }

    except:
        # If anything goes wrong, return empty dictionary
        return {}

    return result
input_str = "Ram|101|22|78, 88, 91"
print(parse_student_record(input_str))
````

#### Question 11: Type Stability and Conversion
Given the list:
```python
data = [10, "20", 30.5, "abc", None, True]
```

Write a function that:
1. Separates numeric and non-numeric values  
2. Converts numeric values to float  
3. Returns both lists  

Explain in comments why `True` behaves as a numeric value.


````python
def separate_numeric(data):
    numeric = []
    non_numeric = []

    for item in data:
        try:
            # True behaves like 1 because in Python bool is a subclass of int
            # So True == 1 and False == 0
            num = float(item)
            numeric.append(num)
        except:
            non_numeric.append(item)

    return numeric, non_numeric

data = [10, "20", 30.5, "abc", None, True]
nums, others = separate_numeric(data)

print("Numeric values:", nums)
print("Non-numeric values:", others)
````

#### Question 13: Set-Based Enrollment Analysis
Given:
```python
python = {"Amit", "Sita", "Hari", "Nita"}
sql = {"Hari", "Ram", "Nita"}
statistics = {"Amit", "Hari", "Gita"}
```

Find:
1. Students enrolled in exactly two subjects  
2. Students enrolled in only one subject  
3. Students enrolled in all subjects


````python
python = {"Amit", "Sita", "Hari", "Nita"}
sql = {"Hari", "Ram", "Nita"}
statistics = {"Amit", "Hari", "Gita"}

# Students enrolled in all subjects
all_subjects = python & sql & statistics

# Students enrolled in exactly two subjects
exactly_two = (
    (python & sql) |
    (python & statistics) |
    (sql & statistics)
) - all_subjects

# Students enrolled in only one subject
only_one = (python | sql | statistics) - exactly_two - all_subjects

print("Students enrolled in all subjects:", all_subjects)
print("Students enrolled in exactly two subjects:", exactly_two)
print("Students enrolled in only one subject:", only_one)
````

#### Question 14: Dictionary Aggregation Logic
Given:
```python
records = {
    "Amit": [78, 82, 91],
    "Sita": [88, 79, 85],
    "Hari": [45, 52, 60]
}
```

Write a function that:
1. Computes average per student  
2. Identifies top and bottom performers  
3. Returns results using a single dictionary


````python
def analyze_records(records):
    result = {}

    # Step 1: find average marks of each student
    for name in records:
        marks = records[name]
        avg = round(sum(marks) / len(marks), 2)
        result[name] = avg

    # Step 2: find top and bottom performer
    top_student = max(result, key=result.get)
    bottom_student = min(result, key=result.get)

    # Step 3: store everything in one dictionary
    final_result = {
        "averages": result,
        "top_performer": top_student,
        "bottom_performer": bottom_student
    }

    return final_result

records = {
    "Amit": [78, 82, 91],
    "Sita": [88, 79, 85],
    "Hari": [45, 52, 60]
}

print(analyze_records(records))
````

#### Question 15: Nested Data Structure Analysis
Given a nested dictionary representing a school's grade book:
```python
school_data = {
    "Class A": {
        "Amit": [78, 82, 91],
        "Sita": [88, 79, 85]
    },
    "Class B": {
        "Hari": [45, 52, 60],
        "Ram": [92, 88, 95]
    }
}
```

Write a function that:
1. Calculates the average marks for each student
2. Finds the top performer in each class
3. Calculates the overall school average
4. Returns a dictionary with:
   - `class_toppers`: {class_name: (student_name, average)}
   - `school_average`: float
   - `student_averages`: {student_name: average}

Explain in comments why nested dictionaries are useful for hierarchical data.

````python
def analyze_school_data(school_data):
    class_toppers = {}
    student_averages = {}
    total_marks = 0
    total_subjects = 0

    for class_name in school_data:
        highest_avg = 0
        topper_name = ""

        for student_name in school_data[class_name]:
            marks = school_data[class_name][student_name]
            avg = round(sum(marks) / len(marks), 2)

            # Store student average
            student_averages[student_name] = avg

            # Update class topper
            if avg > highest_avg:
                highest_avg = avg
                topper_name = student_name

            # Accumulate for school average
            total_marks += sum(marks)
            total_subjects += len(marks)

        # Store topper for the class
        class_toppers[class_name] = (topper_name, highest_avg)

    # Calculate overall school average
    school_average = round(total_marks / total_subjects, 2)

    return {
        "class_toppers": class_toppers,
        "school_average": school_average,
        "student_averages": student_averages
    }
school_data = {
    "Class A": {
        "Amit": [78, 82, 91],
        "Sita": [88, 79, 85]
    },
    "Class B": {
        "Hari": [45, 52, 60],
        "Ram": [92, 88, 95]
    }
}

print(analyze_school_data(school_data))
````
